# create dataframe using list

```python
name=['ramesh','suresh','sathish']
age=[30,40,50]
name,age
```

```python
pd.DataFrame(zip(name,age))
```

$provide$ $columns$

```python

data=zip(name,age)
cols=['name','age']
pd.DataFrame(data,columns=cols)
#pd.DataFrame(data,columns=["name","age"])

```

```python

data=zip(name,age)
cols=['name','age']
ind=['A','B','C']
pd.DataFrame(data,columns=cols,
            index=ind)
#df=pd.DataFrame(data,columns=["name","age"],index=["a","b","c"])
```

# step-5
$add$ $new$ $column$

```python
name=['ramesh','suresh','sathish']
age=[30,40,50]
data=zip(name,age)
cols=['name','age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
df
```

- if u want to add new column
- df['new column']
- you need to have a list which having some elements 
- that elements need to be equal to number of rows
- city_names=['hyd','blr','chennai']
- df['city'=city_names]


```python
city_names=['hyd','blr','chennai']
df['city']=city_names
df
```

# step-6
$update$ $the$ $existing$ $columns$
- if u want to create new coloumn or  update the old column
- both are same way

```python
df['name']=['arun','vamshi','chakry']
df
```

***step-7***

$drop$ $the$ $column$
- in order to drop the column
- we need to use drop method 
- it takes three parameters
- drop the column or row
- mention the column name
- axis=1 reference as column
- axis=0 refernce as row
- if u want to create a newdataframe or 
- u want to overwrite the existing dataframe-inplace=true

```python
df.drop('city', # column name
axis=1, # Column
inplace=True) # overwrite the same
```

```python
df
```

```python
name=['ramesh','suresh','sathish']
age=[30,40,50]
df=pd.DataFrame(zip(name,age),
columns=['name','age'],
index=['A','B','C'])
city_names=['hyd','blr','chennai']
df['city']=city_names
df.drop('city', # column name
axis=1, # Column
inplace=True) # overwrite the same
df
```

# step-8
$drop$ $rows$

```python
df.drop('C', # column name
axis=0, # rows
inplace=True) # overwrite the same
df
```

# step-9
$save$ $the$ $dataframes$

```python
df.to_csv("output.csv")
# while saving index consider as extra column
df.to_excel("output.xlsx")
```
