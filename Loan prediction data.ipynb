***Outlier Analysis***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path=r"C:\Users\dell\Data science\Hackathon files\train_ctrUa4K.csv"
loan_df=pd.read_csv(path)
loan_df.head()
```

- steps in outlier analysis

- step-1: find Q1 ,Q2, Q3
- np.percentile (column data ,q)
- step-2 : caluculate the IQR
- IQR=Q3-Q1
- Step-3: lower boundry and upper boundary
- lb: Q1-1.5IQR
- ub : Q3+1.5IQR
- STEP- 4 : find the outliersdf c1:column data lb c: apply the main condition
- main data [c]

```python
q1=np.percentile(loan_df['ApplicantIncome'],25)
q2=np.percentile(loan_df['ApplicantIncome'],50)
q3=np.percentile(loan_df['ApplicantIncome'],75)
###########################step2##########################
IQR=q3-q1
############################step-3########################
lb=q1-1.5*IQR
ub=q3+1.5*IQR
########################step 4###########################
c1=loan_df['ApplicantIncome']<lb
c2=loan_df['ApplicantIncome']>ub
con=c1|c2
##################step-5################################
outliers_df=loan_df[con]
outliers_df
#######################step- 6##########################
#non_outliers_df=loan_df[c1 & c2]
#non_outliers_df
```

```python
q1=np.percentile(loan_df['ApplicantIncome'],25)
q2=np.percentile(loan_df['ApplicantIncome'],50)
q3=np.percentile(loan_df['ApplicantIncome'],75)
###########################step2##########################
IQR=q3-q1
############################step-3########################
lb=q1-1.5*IQR
ub=q3+1.5*IQR
########################step 4###########################
c1=loan_df['ApplicantIncome']>lb
c2=loan_df['ApplicantIncome']<ub
con=c1&c2
##################step-5################################
non_outliers_df=loan_df[c1 & c2]
non_outliers_df
```

```python
c1 = loan_df['Education'] == 'Graduate'
c2 = loan_df['Loan_Status'] == 'Y'
c3 = loan_df['Loan_Status'] == 'N'

Y = c1 & c2
N = c1 & c3

Y_count = len(loan_df[Y])
N_count = len(loan_df[N])

print(f"There are {Y_count} Graduates with approved loans.")
print(f"There are {N_count} Graduates with denied loans.")
```

```python
labels = loan_df['Education'].unique()
Graduate_count = []
Non_Graduate_count = []

for i in labels:
    c1 = loan_df['Education'] == i
    c2 = loan_df['Loan_Status'] == 'Y'
    c3 = loan_df['Loan_Status'] == 'N'
    Graduate_con = c1 & c2
    Non_Graduate_con = c1 & c3
    Graduate_count.append(len(loan_df[Graduate_con]))
    Non_Graduate_count.append(len(loan_df[Non_Graduate_con]))

cols = ['Education', 'Graduate', 'Non_Graduate']
d1 = pd.DataFrame(list(zip(labels, Graduate_count, Non_Graduate_count)), columns=cols)
d1.set_index('Education', inplace=False)
```
