**categorical vs categorical **

```python
#continents
#case_status
# as we know that there are 25480 observation are there 
# in that 16k are from asia application
# out of 16k applicant how many are approved

#out of 16k applicant how many are denied
```

```python
c1=visa_df['continent']=='Asia'
c2=visa_df['case_status']=='Certified'
c3=visa_df['case_status']=='Denied'
certified_con=c1&c2
denied_con=c1&c3
certified_count=len(visa_df[certified_con])
denied_count=len(visa_df[denied_con])
print(f"there are {certified_count} got certified from Asia")
print(f"there are {denied_count} got denied from Asia")
```

```python
#step-1 : make unique labels
labels=visa_df['continent'].unique()
#step-2 : create empty two lists
certified_count=[]
denied_count=[]
#step-3 : iterate thriugh loop
for i in labels:
    c1=visa_df['continent']== i
    c2=visa_df['case_status']=='Certified'
    c3=visa_df['case_status']=='Denied'
    certified_con=c1&c2
    denied_con=c1&c3
    certified_count.append(len(visa_df[certified_con]))
    denied_count.append(len(visa_df[denied_con]))
cols=['continent','Certified','Denied']

d1=pd.DataFrame(zip(labels,certified_count,denied_count),columns=cols)
d1.set_index('continent')
```
